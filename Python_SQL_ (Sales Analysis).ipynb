```python
#Extract file from zip file
import zipfile as z
var= z.ZipFile('orders.csv.zip')
var.extractall() #extract file to the dir
```

```python
#Read data from the csv files
import pandas
orders_table= pandas.read_csv('orders.csv', na_values = ['Not Available','unknown'])
orders_table.head(10)
```

```python
#handle null values
orders_table['Ship Mode'].unique()
```

```python
#Read column name.. make them lower case and replace score with underscore
#orders_table.rename=({'Order Id':'order id','City':'city'})
orders_table.columns=orders_table.columns.str.lower()
orders_table.columns =orders_table.columns.str.replace(' ','_')
orders_table.head(3)
```

```python
#Assigning orders_table to small variable as t1 .Then derive new columns discounts,sales price and profit
t1=orders_table
t1['discount']= t1['discount_percent']*t1['list_price']/100
t1['sales_price']= t1['list_price']-t1['discount']
t1['profit']=t1['sales_price']-t1['cost_price']
t1.head(3)
```

```python
#Convert order date from object datatype to datetime
t1.dtypes #checking the data types of all the columns
t1['order_date']=pandas.to_datetime(t1['order_date'],format="%Y-%m-%d")
t1.dtypes
```

```python
#drop cost price, list price and discount percent column
t1=t1.drop(columns =['cost_price','list_price','discount_percent'])
t1.head(3)
```
